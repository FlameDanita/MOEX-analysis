```python
import apimoex
import requests
import datetime
import pandas as pd

from matplotlib import pyplot as plt
```

```python
ticker = 'SBER'
interval = 60

end_date = datetime.datetime.now().date()
start_date = end_date - datetime.timedelta(days=365)

# start_date = '2023-01-01'
# end_date = '2024-01-01'
# end_date = start_date + datetime.timedelta(days=5)
```

```python
# Размер свечки - целое число 1 , 10, 60, 24, 7, 31 или 4 (1 квартал)
data = pd.DataFrame(
    apimoex.get_market_candles(
        requests.Session(),
        security=ticker,
        interval=interval,
        start=str(start_date),
        end=str(end_date)))
```

```python
data.tail()
```

```python
data.to_csv(str(ticker) + "_" + str(interval) + "_" + str(start_date) + "_" + str(end_date) + ".csv", index=False)
```
